## **Python Condition If Statements**

**If**

```python
x = 6969
z = 9696
if x < z:
  print ("x is less than z")
```

**If - Elif**

```python
x = 52
z = 51
if x <= z:
  print ("x is less than or equal to z")
elif x > z:
   print ("okay") 
```

**If - Elif - Else**

```python
x = 51
z = 51
if x < z:
  print ("x is less than or equal to z")
elif x > z:
   print ("okay") 
else:
  print ("salamander")
```

**If - Else**

```python
a = 12
b = 8
if a <= b:
  print ("a is less than or equal to b")
else:
  print ("JHBVJDB")
```

**Leap Year**

```python
year= eval(input("Please input the year"))

if ((year%400==0) or (year%4==0 and year%100!=0)):
  print("{0} is a leap year".format(year))
else:
  print("{0} is not a leap year".format(year))
```

**Break**

```python
for x in range(69):
  print (x)
  if x > 5:
    break
```

**Continue**

```python
for x in range(10):
  if x == 5:
    continue 

  print (x)
```

**While loop**

```python
i = 2
while i < 16:
  print(i)
  i += 2
```
